```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("https://www.dataquest.io/wp-content/uploads/2019/09/amazon-orders.csv")
df.head()
```

```python
df.shape
```

```python
df = df.fillna(0)
df.head()
```

```python
df['Total Charged'] = df['Total Charged'].str.replace('$','').astype(float)
df.head()
```

```python
moneyspent = df['Total Charged'].sum()
print(moneyspent)
```

```python
maxspend = max(df['Total Charged'])
print(maxspend)
```

```python
minspend = min(df['Total Charged'])
print(minspend)
```

```python
avgspend = np.mean(df['Total Charged'])
print(avgspend)
```

```python
medspend = np.median(df['Total Charged'])
print(medspend)
```

```python
df['Tax Charged'] = df['Tax Charged'].str.replace('$','').astype(float)
df.head()
```

```python
tax = df['Tax Charged'].sum()
print(tax)
```

```python
taxrate = (tax/moneyspent)*100
print(taxrate)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
plt = sns.barplot(x = df['Order Date'],y = df['Total Charged'])
```

```python
df['year'] = df['Order Date'].dt.year
df['month'] = df['Order Date'].dt.month
df['day']  = df['Order Date'].dt.day
df.head()
```

```python
plt = sns.barplot(x = df['year'],y = df['Total Charged'])
```

```python
plt = sns.barplot(x = df['month'],y = df['Total Charged'])
```

```python
plt = sns.barplot(x = df['day'],y = df['Total Charged'])
```

```python
dailyorders = df.groupby('day').sum()['Total Charged']
dailyorders
```
